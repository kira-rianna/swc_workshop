```python
import glob

files = sorted(glob.glob('../data/chlorophyll-*.csv'))
i = 1

for f in files:
    data = numpy.loadtxt(f, delimiter=',')

    ave_profile = numpy.mean(data, axis=0)
    std_profile = numpy.std(data, axis=0)
    max_profile = numpy.max(data, axis=0)
    min_profile = numpy.min(data, axis=0)

    fig = plt.figure(figsize=(8, 8))
    ax1 = fig.add_subplot(221)
    ax1.plot(ave_profile, depth, label='Mean Cl')
    plt.legend()
    ax2 = fig.add_subplot(222)
    ax2.plot(std_profile, depth, label='Std Cl')
    plt.legend()
    ax3 = fig.add_subplot(223)
    ax3.plot(max_profile, depth, label='Max Cl')
    plt.legend()
    ax4 = fig.add_subplot(224)
    ax4.plot(min_profile, depth, label='Min Cl')
    plt.legend()
    plt.suptitle(f, fontsize=18)
    plt.savefig('../figs/chlorophyll_stats_%s.png' %i, dpi=150)
    i = i+1
```
